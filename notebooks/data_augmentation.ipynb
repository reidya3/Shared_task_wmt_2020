```python
en_de_train_df =  pd.read_csv("../data/en-de/train.ende.df.short.tsv",sep="\t")
en_de_train_df
```

There are some erros in the training dataset where the values for scores and means are mixed up.

```python
mean_561 = en_de_train_df.loc[561,"mean"] 
en_de_train_df.loc[561,"mean"] = en_de_train_df.loc[561,"scores"]
en_de_train_df.loc[561,"scores"] = mean_561

mean_841 = en_de_train_df.loc[842,"mean"]
en_de_train_df.loc[842,"mean"] = en_de_train_df.loc[842,"scores"] 
en_de_train_df.loc[842,"scores"]  = mean_841

mean_1263 = en_de_train_df.loc[1263,"mean"]
en_de_train_df.loc[1263,"mean"] = en_de_train_df.loc[1263,"scores"] 
en_de_train_df.loc[1263,"scores"] = mean_1263
                          
mean_3592 = en_de_train_df.loc[3592,"mean"]                      
en_de_train_df.loc[3592,"mean"] = en_de_train_df.loc[3592,"scores"]
en_de_train_df.loc[3592,"scores"] = mean_3592
                          
mean_3646 = en_de_train_df.loc[3646,"mean"] 
en_de_train_df[3646,"mean"] =  en_de_train_df.loc[3646,"scores"]
en_de_train_df.loc[3646,"scores"] = mean_3646    
    
mean_3636 = en_de_train_df.loc[3636,"mean"]
en_de_train_df.loc[3636,"mean"] = en_de_train_df.loc[3636,"scores"]
en_de_train_df.loc[3636,"scores"] = mean_3636 
      
mean_3706 =   en_de_train_df.loc[3706,"mean"]  
en_de_train_df.loc[3706,"mean"] = en_de_train_df.loc[3706,"scores"]
en_de_train_df.loc[3706,"scores"] = mean_3706
                          
mean_4534 = en_de_train_df.loc[4534,"mean"]
en_de_train_df.loc[4534,"mean"] = en_de_train_df.loc[4534,"scores"]
en_de_train_df.loc[4534,"scores"] = mean_4534
                          
mean_4863 = en_de_train_df.loc[4863,"mean"] 
en_de_train_df.loc[4863,"mean"] = en_de_train_df.loc[4863,"scores"]
en_de_train_df.loc[4863,"scores"] = mean_4863

mean_4863 = en_de_train_df.loc[4863,"mean"]                         
en_de_train_df.loc[4863,"mean"] = en_de_train_df.loc[4863,"scores"]
en_de_train_df.loc[4863,"scores"] = mean_4863
```

```python
en_de_train_df[["mean"]] = en_de_train_df[["mean"]].apply(pd.to_numeric)
```

```python
augmented_df = en_de_train_df[en_de_train_df["mean"]>=90].reset_index(drop=True)
```

## Google Translate Webscraper

```python
driver = webdriver.Chrome()
driver.set_page_load_timeout(30)
lang_code = "de" # german
```

```python
for index, text_input in augmented_df["original"].items():
    time.sleep(1)
    driver.get("https://translate.google.com/#view=home&op=translate&sl=en&tl="+lang_code)
    time.sleep(5 + random.uniform(0,2))
    driver.find_element_by_tag_name("textarea").send_keys(text_input)
    time.sleep(7)
    output_text = driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[1]/div[2]/div[1]/div[1]/div[2]/div[3]/div[1]/div[2]/div/span[1]/span").text
    augmented_df.loc[index, "translation"] = output_text
    time.sleep(2)
```

```python
augmented_df.to_csv("../augmented-data/en-de-augmented-data.tsv", index = None, header=True,sep = '\t')
```
